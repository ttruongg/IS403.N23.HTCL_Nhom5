# <h1> 1. Import libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
import time
```

<h2> Read file csv

```python
df = pd.read_csv('Clean_Data.csv')
df
```

```python
df.info() 
```

<h2> Drawing heatmap chart to see overview of the dataset

```python
plt.figure(figsize=(16,9))
sns.heatmap(df.corr(method='pearson'),annot=True)
```

# <h1>2. Split the data into two parts, one containing the normal attributes, the other containing the decision attribute alone(Price).

```python
X = df.drop(['MODEL_NAME', 'PRICE'], axis=1)
Y = df['PRICE']
```

###Convert non-numeric columns to one-hot vectors to match library input.

Define non-numeric properties in the features

```python
X.select_dtypes(exclude=['int', 'float']).columns
```

Combine pandas library get_dummies statement to convert these columns to one-hot vector

```python
X_onehot = pd.get_dummies(X,columns=X.select_dtypes(exclude=['int64']).columns)
X_onehot
```

# <h1> 3.Split the data lines into two parts training and testing with a ratio of 7-3

```python
X_train,X_test,y_train,y_test = train_test_split(X_onehot,Y,test_size=0.3,random_state=42)
```

# <h1> 4. Lasso Regression Implementation

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.linear_model import Lasso
```

```python
# Lasso Regression model initialization.
lasso = Lasso()

# Setting the alpha values for experimentation.
alphas = [0.01, 0.1, 1, 10, 100]

# Setting up the searching grid for parameter.
param_grid = {'alpha': alphas}

# Using GridSearchCV to finding the best suitable anpha values.
grid_search = GridSearchCV(lasso, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Choose the best anpha values
best_alpha = grid_search.best_params_['alpha']
print("Best Alpha:", best_alpha)
```

```python
# Initializing the Lasso model with optimal alpha.
lasso_best = Lasso(alpha=best_alpha)

start_time_Lasso = time.time()

lasso_best.fit(X_train, y_train)

end_time_Lasso = time.time()
```

```python
running_time_Lasso = end_time_Lasso - start_time_Lasso
print("The running time of the model:", running_time_Lasso, "second")
```

```python
y_pred_Lasso = lasso_best.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2_Lasso = r2_score(y_test, y_pred_Lasso)
print("R-squared:", r2_Lasso)
```

```python
# Determining the sequence number.
x = range(len(y_test))
# Drawing line plot diagram
plt.plot(x, y_test,  label='Thực tế')
plt.plot(x, y_pred_Lasso,  label='Dự đoán')
plt.xlabel('Laptop')
plt.ylabel('Giá')
plt.title('So sánh giá thực tế và giá dự đoán')
plt.legend()
plt.show()
```

# <h1> 5.SVR Implementation

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

# The kernels for experimentation.
kernels = ['linear', 'poly', 'rbf', 'sigmoid']

# The hyperparameters of each kernel for searching.
param_grid = {
    'linear': {'C': [0.1, 1, 10]},
    'poly': {'C': [0.1, 1, 10], 'degree': [2, 3, 4]},
    'rbf': {'C': [0.1, 1, 10], 'gamma': [0.01, 0.001, 0.0001]},
    'sigmoid': {'C': [0.1, 1, 10], 'gamma': [0.01, 0.001, 0.0001]}
}

# Iterating through each kernel and searching for the best hyperparameters for them.
best_scores = {}
best_params = {}
for kernel in kernels:
    svr = SVR(kernel=kernel)
    grid_search = GridSearchCV(svr, param_grid[kernel], cv=5, scoring='neg_mean_squared_error')
    grid_search.fit(X_train, y_train)
    best_scores[kernel] = grid_search.best_score_
    best_params[kernel] = grid_search.best_params_

# Print out the best result for each kernel
for kernel in kernels:
    print(f"Best score for {kernel} kernel: {best_scores[kernel]}")
    print(f"Best parameters for {kernel} kernel: {best_params[kernel]}\n")
```

```python
svr = SVR(kernel='linear', C=10)

start_time_SVR = time.time()

# Training model base on training data
svr.fit(X_train, y_train)

end_time_SVR = time.time()

# Predicting values on testing dataset
y_pred_SVR = svr.predict(X_test)

running_time_SVR = end_time_SVR - start_time_SVR
print("The running time of the model:", running_time_SVR, "second")

from sklearn.metrics import r2_score

r2_SVR = r2_score(y_test, y_pred_SVR)
print("R-squared:", r2_SVR)
```

```python
# Determining the sequence number.
x = range(len(y_test))
# Drawing line plot diagram
plt.plot(x, y_test,  label='Thực tế')
plt.plot(x, y_pred_SVR,  label='Dự đoán')
plt.xlabel('Laptop')
plt.ylabel('Giá')
plt.title('So sánh giá thực tế và giá dự đoán')
plt.legend()
plt.show()
```
